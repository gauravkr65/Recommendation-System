```python
# Select only the specified columns
selected_columns = ['genres', 'keywords', 'overview', 'title', 'movie_id', 'cast', 'crew']
final = merged[selected_columns].copy()
```

```python
# Function to extract names from the list of dictionaries
def extract_names(column_str, key='name', limit=None, filter_by=None, filter_value=None):
    try:
        items = ast.literal_eval(column_str)
        if filter_by and filter_value:
            items = [item for item in items if item.get(filter_by) == filter_value]
        names = [item[key] for item in items]
        if limit:
            names = names[:limit]
        return names
    except (ValueError, SyntaxError):
        return []
```

```python
# Apply the function to the relevant columns using .loc to avoid SettingWithCopyWarning
final.loc[:, 'genres'] = final['genres'].apply(lambda x: extract_names(x, 'name'))
final.loc[:, 'keywords'] = final['keywords'].apply(lambda x: extract_names(x, 'name'))
final.loc[:, 'cast'] = final['cast'].apply(lambda x: extract_names(x, 'name', limit=3))  # Limit to first three names
final.loc[:, 'crew'] = final['crew'].apply(lambda x: extract_names(x, 'name', filter_by='job', filter_value='Director'))  # Only Director's name
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
final['tags'] = final['tags'].apply(stem)
```

```python
final['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
text_vector = cv.fit_transform(final['tags']).toarray()
```

```python
text_vector
```

```python
text_vector[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(text_vector)
```

```python
sorted(list(enumerate(similar[0])),reverse = True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    # Convert the input movie title to lower case
    movie = movie.lower()
    
    # Check if the movie exists in the DataFrame
    if movie not in final['title'].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return
    
    # Find the index of the movie in the DataFrame
    movie_index = final[final['title'] == movie].index[0]
    
    # Get the similarity scores for the movie
    distance = similar[movie_index]
    
    # Sort the movies based on similarity scores and get the indices of the top 5 similar movies
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]

    # Print the titles of the recommended movies
    for i in movie_list:
        print(final.iloc[i[0]]['title'])
```

```python
recommend('batman begins')
```
